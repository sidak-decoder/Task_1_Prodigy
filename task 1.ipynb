```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('task1.csv')
data
```

```python
data.isnull().any()
```

```python
data.info()
```

```python
data_new  = data.sort_values("Country Name").reset_index()
data_new
```

```python
count_name = data_new.head(30).loc[:,"Country Name"].tolist()
count_2023 = data_new.head(30).loc[:,"2023"].tolist()
```

```python
plt.bar(count_name , count_2023)
plt.xticks(rotation=90)
plt.title("Population index")
plt.show()
plt.plot(count_name , count_2023 , marker = "o" , color = "green")
plt.xticks(rotation=90)
plt.title("Population index")
plt.show()
```

```python
plt.figure().set_figwidth(30)
plt.bar(data_new.loc[:,"Country Name"] , data_new.loc[:,"2022"]  , color = "green")
plt.xticks(rotation=90)

plt.title("Population index")
plt.show()
```
